```python
import tkinter as tk
from tkinter import messagebox
import random
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

class OTPVerificationApp:
    def __init__(self, root):
        self.root = root
        self.root.title("OTP Verification")
        
        # Calculate screen dimensions
        screen_width = root.winfo_screenwidth()
        screen_height = root.winfo_screenheight()
        
        # Set window dimensions
        window_width = int(screen_width * 0.45)
        window_height = int(screen_height * 0.45)
        
        # Center window on screen
        x_position = int((screen_width - window_width) / 2)
        y_position = int((screen_height - window_height) / 2)
        
        self.root.geometry(f"{window_width}x{window_height}+{x_position}+{y_position}")
        
        # Style configuration
        self.root.configure(bg="#f0f0f0")  # Background color
        
        self.label_email = tk.Label(root, text="Enter your email address:", bg="#f0f0f0", font=("Arial", 12))
        self.label_email.pack(pady=(20, 5))
        
        self.entry_email = tk.Entry(root, width=30, font=("Arial", 12))
        self.entry_email.pack(pady=5)
        
        self.btn_send_otp = tk.Button(root, text="Send OTP", command=self.send_otp, font=("Arial", 12), bg="#4CAF50", fg="white", padx=10, pady=5)
        self.btn_send_otp.pack(pady=10)
        
        self.label_enter_otp = tk.Label(root, text="Enter OTP:", bg="#f0f0f0", font=("Arial", 12))
        self.label_enter_otp.pack(pady=10)
        
        self.entry_otp = tk.Entry(root, width=10, font=("Arial", 12))
        self.entry_otp.pack()
        
        self.btn_verify_otp = tk.Button(root, text="Verify OTP", command=self.verify_otp, font=("Arial", 12), bg="#008CBA", fg="white", padx=10, pady=5)
        self.btn_verify_otp.pack(pady=10)
        
        self.label_status = tk.Label(root, text="", bg="#f0f0f0", font=("Arial", 12))
        self.label_status.pack(pady=10)
        
    def generate_otp(self):
        """Generate a 6-digit OTP."""
        return str(random.randint(100000, 999999))
    
    def send_email_otp(self, receiver_email, otp):
        """Send OTP to the provided email address."""
        sender_email = 'Reshmabegum3804@gmail.com'  # Replace with your email address
        password = 'lrtz ujsg jeiv bnsb'  # Replace with your email password
        
        message = MIMEMultipart()
        message['From'] = sender_email
        message['To'] = receiver_email
        message['Subject'] = 'Your OTP Code'

        body = f"Your OTP is {otp}."
        message.attach(MIMEText(body, 'plain'))

        try:
            server = smtplib.SMTP('smtp.gmail.com', 587)
            server.starttls()
            server.login(sender_email, password)
            text = message.as_string()
            server.sendmail(sender_email, receiver_email, text)
            return True
        except Exception as e:
            print(f"Error sending email: {e}")
            return False
        finally:
            server.quit()
    
    def send_otp(self):
        receiver_email = self.entry_email.get().strip()
        
        if receiver_email:
            otp = self.generate_otp()
            sent = self.send_email_otp(receiver_email, otp)
            
            if sent:
                self.current_otp = otp  # Store the current OTP for verification
                # messagebox.showinfo("OTP Sent", "OTP has been sent to your email.")
                self.label_status.config(text="OTP has been sent to your email.")

            else:
                messagebox.showerror("Error", "Failed to send OTP. Please try again.")
        else:
            messagebox.showerror("Error", "Please enter your email address.")
    
    def verify_otp(self):
        entered_otp = self.entry_otp.get().strip()
        
        if entered_otp == self.current_otp:
            # 
            self.label_status.config(text="Valid OTP")
            messagebox.showinfo("OTP Valid", "OTP is valid. Redirecting to another page.")
            self.root.destroy()

            # Add code here to navigate to another page or perform an action
        else:
            messagebox.showerror("OTP Invalid", "Entered OTP is not valid. Please try again.")
            self.entry_otp.delete(0, tk.END)  # Clear the entry field

if __name__ == "__main__":
    root = tk.Tk()
    app = OTPVerificationApp(root)
    root.mainloop()
```
